# Steam Store Review Prediction

## Data Collection

I have gathered all my data from the steam store

**What is the steam store?**

Its a video game digital distribution service. It contains thousand of games. As of writting this there are currently around 80,000 games currently on the store. However only around 37,000 of these games have a review. These reviews are user submitted. 

**About the data**

As said previously the Reviews are user submitted and can be one of 9 results.
- 95 - 99% : Overhwelmingly Positive
- 94 - 80% : Very Positive
- 80 - 99% + few reviews: Positive
- 70 - 79% : Mostly Positive
- 40 - 69% : Mixed
- 20? - 39% : Mostly Negative
- 0 - 39% + rew reviews: Negative
- 0 - 19% : Very Negative
- 0 - 19% + many reviews: Overwhelmingly Negative

Next the tags. There are hundreads of tags available. I did not know there were that many. I am only taking the top 20 tags of each game.
The rest of the data is self explanatory

**How did I gather the data**

hrefScraper.ipynb: I used a python/selenium scraper to gather the link of every game page on the steam store that has a review. 



Scraper.ipynd: I made another scraper that went to each page to retreive all the info. That included the game tags, the review, name, developer name and the date of the release.



I decided to make these two scraper seperate because it made it easier for me to test both at the same time. Gathering the Href took about 12 hours. I split the href data in two table and used two computers to gather the data. Both running concurrently ran for around 36 hours. I then had to merge the two df back togheter.
